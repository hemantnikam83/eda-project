```python
## IMPORTING THE LIBRARIES:-
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## LOADING THE DATA
df= pd.read_csv("/content/crime_dataset_india.csv")
df
```

```python
#DESCRIBING THE DATA
df.describe()
```

```python
#INFORMATION OF DATA
df.info()
```

```python
#HANDALING MISING DATA:-
df.dropna()
```

```python
#EXTRACTING YEAR FROM DATE
df['Year Of Occurance'] = pd.to_datetime(df['Date of Occurrence']).dt.year
df
```

```python
#GROUPING DATA BY CRIME_DESCRIPTION & CITY
crime_summary = df.groupby(['Year Of Occurance','Crime Description','City']).size().reset_index(name='Count')
crime_summary
```

```python
#MEAN OF CRIME
mean_crime_rate=np.mean(crime_summary['Count'])
#STANDARD DEVIATION
std_crime_rate=np.std(crime_summary['Count'])
print(f"mean crime rate: {mean_crime_rate}")
print(f"standard deviation of crime rate: {std_crime_rate}")
```

```python
# Gender wise crime
GenderwiseCrime=df.groupby('Victim Gender')['Crime Description'].value_counts()
GenderwiseCrime
```

```python
#MEAN AND STANDARD DEVIATION
mean_crime_gender=np.mean(GenderwiseCrime)
std_crime_gender=np.std(GenderwiseCrime)
print(f"mean crime rate: {mean_crime_gender}")
print(f"standard deviation of crime rate: {std_crime_gender}")
```

```python
#VISUALISATION
yearly_trend=crime_summary.groupby('Year Of Occurance')['Count'].sum()
plt.figure(figsize=(10,6))
plt.plot(yearly_trend.index,yearly_trend.values,marker='*',linestyle='--')
plt.title('Crime Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.grid()
plt.show()
```

```python
#Gender wise crime trend over the years
gender_yearly_trend = df.groupby(['Year Of Occurance', 'Victim Gender']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 7))
for gender in gender_yearly_trend.columns:
    plt.bar(gender_yearly_trend.index, gender_yearly_trend[gender], label=gender)

plt.title('Crime Trends Over the Years by Gender')
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.legend(title='Victim Gender')
plt.show()
```
